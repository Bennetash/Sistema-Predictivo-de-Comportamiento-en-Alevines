```python
import os
os.environ["QT_QPA_PLATFORM"]= "xcb"

from ultralytics import YOLO
from ultralytics.solutions import heatmap
import cv2
import csv
import datetime

url = 'http://192.168.1.25:4747/video'
cap = cv2.VideoCapture(url)

if not cap.isOpened():
    print('Error video stream')
    exit(0)

model = YOLO('best.pt')

imw = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
imh = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

heatmap_obj = heatmap.Heatmap()
heatmap_obj.set_args(colormap=cv2.COLORMAP_JET,
                     imw=imw,
                     imh=imh,
                     heatmap_alpha=0.35,
                     region_thickness=2,
                     view_img=True,)

csv_file = 'detecciones_alevines.csv'
if not os.path.exists(csv_file):
    with open(csv_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Timestamp', 'Cantidad_Alevines', 'Tiempo_Inferencia_ms', 'x', 'y', 'w', 'h'])


while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        break

    im0_resized = cv2.resize(im0, (imw, imh))
    results = model.track(im0_resized, persist=True, conf = 0.6)
    frame = heatmap_obj.generate_heatmap(im0_resized, tracks=results)

timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
cantidad_alevines = len(results.xyxy[0])
tiempo_inferencia = sum(results.times)*1000

for i in range(cantidad_alevines):
    x, y, w, h = results.xyxy[0][i][:4]
    with open(csv_file, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([timestamp, cantidad_alevines, tiempo_inferencia, x, y, w, h])

cap.release()
cv2.destroyAllWindows()
```
