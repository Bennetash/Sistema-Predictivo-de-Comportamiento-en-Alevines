```python
import os
import cv2
import csv
import datetime
from ultralytics import YOLO
from ultralytics.solutions import heatmap

def process_and_save_detections(results, timestamp, csv_file):
    if results.boxes is not None:
        for box in results.boxes.data:
            x1, y1, x2, y2, conf, cls_id = box[:6]
            cls_name = results.names[int(cls_id)]
            with open(csv_file, mode='a') as file:
                writer = csv.writer(file)
                writer.writerow([timestamp, cls_name, conf, x1, y1, x2, y2])

def inicializador_guardado_video(video_index):
    return cv2.VideoWriter(f'monitoreo_alevines_output_{video_index}.mp4',
                           cv2.VideoWriter_fourcc(*'mp4v'),
                           30.0,
                           (imw, imh))

os.environ["QT_QPA_PLATFORM"]= "xcb"
url = 'http://192.168.1.25:4747/video'
cap = cv2.VideoCapture(url)
model = YOLO('best.pt')
imw = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
imh = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
max_duration = 5 * 60
max_frames = fps * max_duration
frame_count = 0
video_index = 1
heatmap_obj = heatmap.Heatmap()
csv_file = 'detecciones_alevines.csv'

heatmap_obj.set_args(colormap=cv2.COLORMAP_JET,
                     imw=imw,
                     imh=imh,
                     heatmap_alpha=0.35,
                     region_thickness=2,
                     view_img=True,)

if not os.path.exists(csv_file):
    with open(csv_file, mode='w') as file:
        writer = csv.writer(file)
        writer.writerow(['timestamp', 'class', 'confidence', 'X1', 'Y1', 'X2', 'Y2'])

video_writer = inicializador_guardado_video(video_index)

while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Frames de video sin objeto o video procesado de manera exitosa")
        break

    if frame_count == max_frames:
        video_writer.release()
        video_index += 1
        video_writer = inicializador_guardado_video(video_index)
        frame_count = 0

    timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    im0_resized = cv2.resize(im0, (imw, imh))
    results = model.track(im0_resized, persist=True, conf = 0.6)
    frame = heatmap_obj.generate_heatmap(im0_resized, tracks=results)

    video_writer.write(frame)
    frame_count += 1

    #print(results)

    if isinstance(results, list):
        for results in results:
            process_and_save_detections(results, timestamp, csv_file)
    else:
        process_and_save_detections(results, timestamp, csv_file)

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
