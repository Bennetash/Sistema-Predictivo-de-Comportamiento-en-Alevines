```python
import os
os.environ["QT_QPA_PLATAFORM"]= "xcb"

from ultralytics import YOLO
from ultralytics.solutions import heatmap
import cv2

model = YOLO("yolov8s.pt")
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error opening video stream or file")
    exit(0)

imw = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
imh = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

heatmap_obj = heatmap.Heatmap()
heatmap_obj.set_args(colormap = cv2.COLORMAP_JET,
                     imw = imw,
                     imh = imh,
                     view_img = True,)
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        exit(0)
    
    im0_resized = cv2.resize(im0, (imw, imh))

    results = model.track(im0_resized, persist=True)
    frame = heatmap_obj.generate_heatmap(im0_resized, tracks = results)
```
